```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
report2018_file = "./WorldHappiness2018_Data.csv"
report2018_df = pd.read_csv(report2018_file)
report2018_df.head()
```

```python
report2017_file = "./2017.csv"
report2017_df = pd.read_csv(report2017_file)
report2017_df.head()
```

```python
# Create a filtered dataframe from specific columns
report2017_cols = ["Country", "Happiness.Rank", "Happiness.Score", "Economy..GDP.per.Capita.", 
                   "Health..Life.Expectancy.", "Freedom", "Generosity", "Trust..Government.Corruption."]

report2017_transformed= report2017_df[report2017_cols].copy()

# Rename the column headers
report2017_transformed= report2017_transformed.rename(columns={"Happiness.Rank": "rank", "Happiness.Score": "score", 
                                         "Economy..GDP.per.Capita.": "gdp_per_capita", 
                                         "Health..Life.Expectancy.": "health_life_expectancy",
                                         "Trust..Government.Corruption.": "corruption",
                                         "Country": "country", "Generosity": "generosity",
                                         "Freedom": "freedom"})

report2017_transformed.head()
```

```python
# Create a filtered dataframe from specific columns
report2018_cols = ["Country", "Rank", "Score", "GDP_Per_Capita", "Generosity", "Social_Support",
                   "Healthy_Life_Expectancy", "Freedom_To_Make_Life_Choices", "Perceptions_Of_Corruption"]

report2018_transformed= report2018_df[report2018_cols].copy()


# Rename the column headers
report2018_transformed= report2018_transformed.rename(columns={"Country": "country", "Score": "score",
                                          "Social_Support": "social_support", "Rank": "rank",
                                           "GDP_Per_Capita": "gdp_per_capita", "Generosity": "generosity",
                                          "Healthy_Life_Expectancy": "health_life_expectancy",
                                          "Freedom_To_Make_Life_Choices": "freedom", 
                                          "Perceptions_Of_Corruption": "corruption"})

report2018_transformed.head()
```

```python
#Create database connection
connection_string = "postgres:sdsndn8891@localhost:5432/happiness_report_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
report2018_transformed.to_sql(name='report2018', con=engine, if_exists='append', index=True)
```

```python
report2017_transformed.to_sql(name='report2017', con=engine, if_exists='append', index=True)
```
